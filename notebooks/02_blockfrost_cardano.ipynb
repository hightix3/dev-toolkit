# Cardano Blockchain Explorer

Query the Cardano blockchain using Blockfrost API + PyCardano.

## Prerequisites
1. Sign up at [blockfrost.io](https://blockfrost.io)
2. Create a **testnet** project
3. Copy the API key to your `.env` file as `BLOCKFROST_PROJECT_ID`

```python
import sys
sys.path.insert(0, '..')

from src.blockchain import CardanoClient, WalletManager
from src.utils import get_settings
```

## 1. Check Connection

```python
client = CardanoClient()

health = client.check_health()
print(f'Blockfrost API healthy: {health}')
```

## 2. Latest Block

```python
block = client.get_latest_block()
print(f'Block height: {block["height"]}')
print(f'Epoch:        {block["epoch"]}')
print(f'Slot:         {block["slot"]}')
print(f'Transactions: {block["tx_count"]}')
print(f'Hash:         {block["hash"][:20]}...')
```

## 3. Network Stats

```python
network = client.get_network_info()

total_ada = int(network['supply_total']) / 1_000_000
circulating_ada = int(network['supply_circulating']) / 1_000_000

print(f'Total supply:       {total_ada:,.0f} ADA')
print(f'Circulating supply: {circulating_ada:,.0f} ADA')
print(f'Active stake:       {int(network["stake_active"]) / 1_000_000:,.0f} ADA')
```

## 4. Generate a Testnet Wallet

```python
wallet = WalletManager(network='testnet')
address = wallet.generate_address()
print(f'New testnet address: {address}')
print(f'\nFund this address using the Cardano testnet faucet:')
print(f'https://docs.cardano.org/cardano-testnets/tools/faucet/')
```
