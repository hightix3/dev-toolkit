```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from src.data_analysis import DataLoader, DataAnalyzer, DataVisualizer

sns.set_theme(style='whitegrid')
%matplotlib inline
```

## 1. Load Data

```python
loader = DataLoader()

# Option A: Load from local file
# df = loader.load_csv('../data/raw/your_dataset.csv')

# Option B: Load from URL
# df = loader.load_from_url('https://example.com/data.csv')

# Option C: Sample dataset for testing
df = pd.DataFrame({
    'date': pd.date_range('2024-01-01', periods=100, freq='D'),
    'value': np.random.randn(100).cumsum() + 100,
    'category': np.random.choice(['A', 'B', 'C'], 100),
    'amount': np.random.exponential(50, 100),
})

print(f'Shape: {df.shape}')
df.head()
```

## 2. Initial Inspection

```python
print('--- Data Types ---')
print(df.dtypes)
print(f'\n--- Shape: {df.shape} ---')
print(f'\n--- Memory: {df.memory_usage(deep=True).sum() / 1024:.1f} KB ---')
```

## 3. Missing Value Analysis

```python
analyzer = DataAnalyzer(df)
missing = analyzer.missing_report()
if len(missing) > 0:
    print(missing)
else:
    print('No missing values found.')
```

## 4. Statistical Summary

```python
df.describe()
```

## 5. Visualizations

```python
viz = DataVisualizer(df)

# Distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
df['value'].hist(bins=25, ax=axes[0], edgecolor='white')
axes[0].set_title('Value Distribution')
df['amount'].hist(bins=25, ax=axes[1], edgecolor='white')
axes[1].set_title('Amount Distribution')
plt.tight_layout()
plt.show()
```

```python
# Time series
fig, ax = plt.subplots(figsize=(12, 5))
ax.plot(df['date'], df['value'])
ax.set_title('Value Over Time')
ax.set_xlabel('Date')
ax.set_ylabel('Value')
plt.show()
```

```python
# Category breakdown
fig, ax = plt.subplots()
sns.boxplot(data=df, x='category', y='amount', ax=ax)
ax.set_title('Amount by Category')
plt.show()
```
